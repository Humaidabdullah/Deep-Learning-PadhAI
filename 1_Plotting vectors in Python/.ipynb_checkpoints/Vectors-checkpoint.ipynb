```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.quiver(0,0,3,4) ## (0,0) to (3,4)
plt.show()
```

```python
#Scale the axes
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1)
plt.show()
```

```python
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1,color="r")
plt.quiver(0,0,-3,4,scale_units='xy',angles='xy',scale=1,color="g")
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Function which takes a list of vectors and plots it.

```python
def plot_vectors(vecs):
    colors = ['r','b','g','y'] #To change colors of the vectors automatically
    i=0
    for vec in vecs:
        plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units='xy',angles='xy',scale=1,color=colors[i%len(colors)])
        i += 1
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.show()
```

```python
vecs = [[0,0,3,4],[0,0,-3,4],[0,0,-3,-4],[0,0,4,-1]]
plot_vectors(vecs)
```

# Addition and Subtraction

```python
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,-3,-4]),np.asarray([0,0,4,-1])]
#We are using numpy to add the vectors
```

```python
plot_vectors([vecs[0],vecs[1]])
```

We're gonna add and plot these two vectors

```python
add_vec = vecs[0]+vecs[1]
```

```python
plot_vectors([vecs[0],vecs[1],add_vec])
```

```python
plot_vectors([-vecs[1]])
```

```python
plot_vectors([vecs[0],vecs[1],vecs[0]-vecs[1]])
```

```python
plot_vectors([vecs[0],vecs[1],vecs[0]+(-vecs[1])])
```

# Dot Product

```python
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,-3,-4]),np.asarray([0,0,4,-1])]
```

```python
plot_vectors(vecs)
```

We are taking vecs[0] and vecs[1]

```python
a = np.asarray([3,4])
b = np.asarray([4,-1]) 
```

$\vec{a}\cdot\vec{b} = |\vec{a}| |\vec{b}| \cos(\theta) $

```python
a_dot_b = np.dot(a,b)
```

```python
a_dot_b
```

## Projection of $\vec{a}$ on $\vec{b}$
$a_b = |\vec{a}| \cos(\theta) = |\vec{a}| \frac{\vec{a} \cdot \vec{b}}{|\vec{a}||\vec{b}|} = \frac{\vec{a} \cdot \vec{b}}{|\vec{b}|}$ 

```python
a_b = np.dot(a,b)/np.linalg.norm(b)
```

```python
a_b
```

```python
# a_b is in the direction of b
```

$\vec{a_b} = a_b \hat{b} = a_b \frac{\vec{b}}{|\vec{b}|}$ $

```python
vec_a_b = (a_b/np.linalg.norm(b))*b
```

```python
print(vec_a_b)
```
